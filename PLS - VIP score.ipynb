```python
#!/usr/bin/env python
# coding: utf-8
import os.path
import warnings
warnings.filterwarnings("ignore")

from scipy.io import loadmat
import numpy as np
from numpy.random import seed
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="ticks", palette='muted')
# import plotly.express as px
```

```python
fontsize = 30
plt.rc('font', size=fontsize*.8) #controls default text size
plt.rc('axes', titlesize=fontsize) #fontsize of the title
plt.rc('axes', labelsize=fontsize) #fontsize of the x and y labels
plt.rc('xtick', labelsize=fontsize/1.5) #fontsize of the x tick labels
plt.rc('ytick', labelsize=fontsize/1.5) #fontsize of the y tick labels
plt.rc('legend', fontsize=fontsize/1.5) #fontsize of the legend
```

```python
data = loadmat("data/pls_vip_scores.mat")
data.keys()
PLS_vip = data['vipScore']
W = data['W'][0]

# PLS_vip[:].shape
```

```python
def plot_variable_importance(coefs, fontsize=30, nbr_of_ticks=10, palette=['#585153']):
    _, ax = plt.subplots(1, figsize=(18,10))
    sns.lineplot(coefs.T, legend=False, ax=ax, alpha=.6, palette=palette)
    # ax.axhline(y=0., color='b', linestyle='--')    
    ax.set(xlabel='Wavenumbers (cm-1)', ylabel='VIP score')
    plt.show()
```

```python
df_PLS_vip = pd.DataFrame(PLS_vip.copy().flatten()).T
df_PLS_vip.columns = W
df_PLS_vip.index = ['Variable importance']

plot_variable_importance(df_PLS_vip, palette=['#888B8D'])
```
